```python
itervalue=list(2*n for n in range(50))     # creation of iter
iterobject=iter(itervalue)   #initialization of iter
while True:
  try:
    items=next(iterobject)
    print(items)
  except StopIteration:
    break
```

```python
#Generator
def calsum(n):
  num=0;
  while num<n:
    yield num             # works like an iterator
    num=num+1
sumofnumber=sum(calsum(10))
print(sumofnumber)
```

```python
# Implementing Simple decorator
def hello_decorator(func):   # defining decorator
  def Inner_Hello():         # This is called as wrapper function within argument is called, It will access outer function too
    print("Hello 1")
    func()                   # calling argument function in wrapper function
    print("Hello 2")
  return Inner_Hello

def Outer_Hello():
  print("Hello 3")

callingfunction=hello_decorator(Outer_Hello) #Passing actual function as parameter of decorator
callingfunction()
```

```python
# Find the execution time of function
import time
import math
def caltime(func):
  def Inner_caltime(*args): # *args in python takes any number of arguments in user-defined function and converts user inputs to a tuple named args.
    begin=time.time()
    func(*args)
    end=time.time()
    print("Total time required for function execution is: ", end-begin)
  return Inner_caltime

@caltime
def factorial(num):
  time.sleep(2)
  print("Factorial of Num is : ", math.factorial(num))
factorial(10)
```

```python
# chaining decorator/Multiple decorators
# d1(d2(outerfunction))
# calculate square of given number

def decor1(func):
  def inner_decor1():
    x=func()
    return x*x
  return inner_decor1

def decor2(func):
  def inner_decor2():
    x=func()
    return 2*x
  return inner_decor2

@decor1
@decor2
def square():
  return 20

print(square())
```
