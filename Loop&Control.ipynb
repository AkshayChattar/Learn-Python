```python
# use of break, continue, pass
#break:It brings control out of the loop
for letter in "welcome to simca":
  if letter =='c':
    break
  print(letter)
```

```python
# use of break, continue, pass
#continue:It brings control out of the loop
for letter in "welcome to simca":
  if letter =='c':
    continue
  print(letter, end= " ")
```

```python
# use of break, continue, pass
#pass:It brings control out of the loop
for letter in "welcome to simca":
  if letter =='c':
    pass
  print(letter, end= " ")
```
